* Create a 4 row and 1 column plot where:
    * The first row is a line graph of all the first quater earnings
    * The second row is a line graph of all the second quater earnings
    * The third row is a line graph of all the third quater earnings
    * The fourth row is a line graph of all the fourth quater earnings
* All the plots should have a title saying what quarter is being shown
* All the plots plots but the final should not have an x-axis, the final plot should only have the even years on the axis
* The y-axis should be called "Earnings"
* The x-axis should be called "Year"
* The whole plot should be titled "Johnson and Johnson Quarterly Earnings"


````python
# Your code below (You need to make the appropriate imports of matplotlib and numpy)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# EOL and maltiple lines
# https://www.techbeamers.com/python-multiline-string/
data = '''
0.71  0.63  0.85  0.44
0.61  0.69  0.92  0.55
0.72  0.77  0.92  0.60
0.83  0.80  1.00  0.77
0.92  1.00  1.24  1.00
1.16  1.30  1.45  1.25
1.26  1.38  1.86  1.56
1.53  1.59  1.83  1.86
1.53  2.07  2.34  2.25
2.16  2.43  2.70  2.25
2.79  3.42  3.69  3.60
3.60  4.32  4.32  4.05
4.86  5.04  5.04  4.41
5.58  5.85  6.57  5.31
6.03  6.39  6.93  5.85
6.93  7.74  7.83  6.12
7.74  8.91  8.28  6.84
9.54 10.26  9.54  8.73
11.88 12.06 12.15  8.91
14.04 12.96 14.85  9.99
16.20 14.67 16.02 11.61
'''
# data turn into a long list
# massage spaces to commas, \n to space AND commas
# data
data_m = np.str.replace(data, '  ', ',')
data_m = np.str.replace(data_m, ' ', ',') # keep working on data_m
data_m = np.str.replace(data_m, '\n', ',') # keep working on data_m
data_m = data_m[1:-1]  # chop off stray "," from head and tail # data_m
str_list = list(data_m.split(",")) #str_list
divid_pQ_Y = [float(ele) for ele in str_list] #float_list = divid_pQ
# prepare X axis for quarters
Qs_X = np.arange(1960, 1981, .25)  # goes to 1980.75
# Qs_X = np.linspace(1960, 1981, 4) # screw linspace need to calculate

Data_pQ = np.reshape(divid_pQ_Y, (-1,4))
# Data_pQ.shape
Ys_X = np.arange(1960, 1981)  # goes to 1980


# Now subplots
fig = plt.figure()
fig.suptitle("Johnson and Johnson Quarterly Earnings")

plt.subplot(4,1,1)
plt.plot(Ys_X, Data_pQ[:,0]) # 1st quarter
plt.title("1st quarter")
plt.xticks([])

plt.subplot(4,1,2)
plt.plot(Ys_X, Data_pQ[:,1]) # 2nd quarter
plt.title("2nd quarter")
plt.xticks([])

plt.subplot(4,1,3)
plt.plot(Ys_X, Data_pQ[:,2]) # 3rd quarter
plt.title("3rd quarter")
plt.xticks([])

plt.subplot(4,1,4)
plt.plot(Ys_X, Data_pQ[:,3]) # 4th quarter
plt.title("4th quarter")
tick_location = Ys_X[1::2]
tick_labels = [str(i) for i in tick_location]
plt.xticks(tick_location, tick_labels)
plt.xlabel("Year")
plt.ylabel("Earnings")
````

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cars_data = pd.read_csv('../data/cars.csv', sep='|')
cars_data.rename(columns={'Unnamed: 0': 'make_model'}, inplace=True)
cars_data.head()
# make_model(0) mpg(1) cyl(2) disp(3) hp(4) drat(5) 
# wt(6) qsec(7) vs am gear carb

# nom_cyl = np.unique(cars_data.loc[:,'cyl'])
# nom_cyl # array([4, 6, 8])
# cyl_grouped = cars_data.groupby('cyl')
# for name, group in cyl_grouped:
#     print(name)
#     print(group)
plt.figure()
    
cyl4_x = cars_data.loc[(cars_data.cyl==4)].iloc[:,4] # extract 'hp'
cyl4_y = cars_data.loc[(cars_data.cyl==4)].iloc[:,7] # extract 'qsec'

plt.subplot(1,3,1)  # 4 cylinders dat
plt.scatter(cyl4_x, cyl4_y)
plt.title("4 cylinders")
plt.xlabel("Horsepower")
plt.ylabel("quarter mile time (qsec)")

cyl6_x = cars_data.loc[(cars_data.cyl==6)].iloc[:,4] # extract 'hp'
cyl6_y = cars_data.loc[(cars_data.cyl==6)].iloc[:,7] # extract 'qsec'

plt.subplot(1,3,2)  # 6 cylinders dat
plt.scatter(cyl6_x, cyl6_y)
plt.title("6 cylinders")
plt.xlabel("Horsepower")
plt.ylabel("quarter mile time (qsec)")


cyl8_x = cars_data.loc[(cars_data.cyl==8)].iloc[:,4] # extract 'hp'
cyl8_y = cars_data.loc[(cars_data.cyl==8)].iloc[:,7] # extract 'qsec'

plt.subplot(1,3,3)  # 8 cylinders dat
plt.scatter(cyl8_x, cyl8_y)
plt.title("8 cylinders")
plt.xlabel("Horsepower")
plt.ylabel("quarter mile time (qsec)")
````
